```python
def classify_petal_width(x):
    if x > 1.2:
        return 'large'
    else:
        return 'small'
iris['PetalWidthSummary'] = iris.PetalWidth.apply(classify_petal_width)
```

```python
def classify_petal_length(x):
    if x > 3.8:
        return 'large'
    else:
        return 'small'
iris['PetalLengthSummary'] = iris.PetalLength.apply(classify_petal_length)
```

```python
def classify_sepal_width(x):
    if x > 3.1:
        return 'large'
    else:
        return 'small'
iris['SepalWidthSummary'] = iris.SepalWidth.apply(classify_sepal_width)
```

```python
def classify_sepal_length(x):
    if x > 5.8:
        return 'large'
    else:
        return 'small'
iris['SepalLengthSummary'] = iris.SepalLength.apply(classify_sepal_length)
```

```python
def classify_flower(row):
    pw = row['PetalWidth']
    pl = row['PetalLength']
    sw = row['SepalWidth']
    sl = row['SepalLength']
    # Hard coded std and means - should fix this
    setosa_score = (((pw -0.24400)**2)/(0.10721**2) + 
                   ((pl - 1.464000)**2)/(0.173511**2) + 
                   ((sw -3.418000)**2)/(0.381024**2) +               
                   ((sl -5.00600)**2)/(0.35249**2))
                
    virginica_score = (((pw -2.02600)**2)/(0.27465**2) + 
                   ((pl - 5.552000)**2)/(0.551895**2) +  
                   ((sw -2.974000)**2)/(0.322497**2) +                 
                   ((sl -6.58800)**2)/(0.63588**2))
    
    versicolor_score = (((pw -1.326000)**2)/(0.197753**2) + 
                   ((pl - 4.260000)**2)/(0.469911**2) +  
                   ((sw -2.770000)**2)/(0.313798 **2) +                 
                   ((sl -5.936000)**2)/(0.516171**2))
    
    
    if (setosa_score < virginica_score and 
        setosa_score < versicolor_score):
        return 'Iris-setosa'
    elif (virginica_score < setosa_score and 
        virginica_score < versicolor_score):
        return 'Iris-virginica'        
    else:
        return 'Iris-versicolor'
```

```python
iris['PredictedName'] = iris.apply(classify_flower,axis=1)
```
